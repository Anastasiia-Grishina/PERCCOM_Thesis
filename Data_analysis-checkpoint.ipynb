### Create pandas with number of processes running at a certain hour

Check if any process has one stage that lasts less than any other stage.

Check if any process has less than 4 stages.

Check if there are process ids that are not incremented by 1.

```python
id_process_prev = lsf_df.loc[0]['id process']
end_time_prev = lsf_df.loc[0]['end_time_py']
start_time_prev = lsf_df.loc[0]['start_time_py']
stage_prev = lsf_df.loc[0]['state proc.']
start_flag = True
end_flag = True

for row_num in range(1, len(lsf_df['id process']) ):
    stage = lsf_df.loc[row_num]['state proc.']
    id_process = lsf_df.loc[row_num]['id process']
    start_time = lsf_df.loc[row_num]['start_time_py']
    end_time = lsf_df.loc[row_num]['end_time_py']
    
    if ( id_process == id_process_prev ):
#         print ( row_num, id_process_prev, id_process, end_time_prev, end_time )
        if ( end_time != end_time_prev ):
            end_flag = False
        if ( start_time != start_time_prev ):
            start_flag = False
    else:
        if (1==stage)and(stage_prev<4):
            print( 'state proc.')
            print( row_num, stage_prev, stage )
        if ( abs(id_process - id_process_prev)>1 ):
            print( 'id process')
            print( row_num, id_process_prev, id_process )
    if not(start_flag):
        print(row_num)
        continue
    if not(end_flag):
        print(row_num)
        continue
    stage_prev = stage
    start_time_prev = start_time
    end_time_prev = end_time 
    id_process_prev = id_process
print(start_flag)
print(end_flag)
```

#### Create 'processes_at_time' data frame where the info from 2 datasets will be combined: 

during the hour reported by Zabbix $x$ processes are working

```python
proc_at_time_df = pd.DataFrame(np.zeros((len(zabbix_df.date_py), 5)), \
                             columns = ['date_py','time_py','processes_running',\
                                        'proc_started_this_hour', 'proc_ended_this_hour'])
proc_at_time_df.date_py = zabbix_df.date_py.copy()
proc_at_time_df.time_py = zabbix_df.time_py.copy()
proc_at_time_df['power_avg']=zabbix_df['power consumption avg'].copy()
proc_at_time_df['power_min']=zabbix_df['power consumption min'].copy()
proc_at_time_df['power_max']=zabbix_df['power consumption max'].copy()
proc_at_time_df.head()
```

```python
i = 0
for lsf_current_id_process in (lsf_df['id process'].unique()):
#     print(i)
    i+=1
    lsf_current_line = lsf_df[lsf_df['id process']==lsf_current_id_process].iloc[0]
#     print(lsf_current_line)
#     print('\n')
    
    #find first line of date and start_time+1 in proc_at_time_df 
    #which are the same as in lsf
    #add +1 process during the reported hour
    #go through time in proc_df 
    #while date<=end_date and end_time.hour (in lsf_df) +1 <= time in proc_df
    
    #did not check zero case 
    lsf_df_start_hour = lsf_current_line.start_time_py.hour
    lsf_df_end_hour = lsf_current_line.end_time_py.hour
    lsf_df_end_date = lsf_current_line.end_date_py
    
    if ( lsf_df_start_hour != 23 ):
        proc_df_current_line_no = proc_at_time_df[ (proc_at_time_df.date_py == lsf_current_line.start_date_py) \
                                              & (proc_at_time_df.time_py > datetime.time(lsf_df_start_hour,0,0)) ].index[0]
    else:
        proc_df_current_line_no = proc_at_time_df[ (proc_at_time_df.date_py > lsf_current_line.start_date_py) \
                                              & (proc_at_time_df.time_py == datetime.time(0,0,0)) ].index[0]
#     print( proc_at_time_df.iloc[proc_df_current_line_no] )
    
    proc_df_hour = proc_at_time_df.loc[proc_df_current_line_no,'time_py'].hour
#     print(proc_df_hour)

    proc_df_date = proc_at_time_df.loc[proc_df_current_line_no,'date_py']
#     print(proc_df_date)
    proc_at_time_df.loc[proc_df_current_line_no,'processes_running'] += 1
    
    if not( (lsf_df_start_hour == lsf_df_end_hour) and (lsf_current_line.start_date_py == lsf_df_end_date) ):
        while ( proc_df_current_line_no < proc_at_time_df.shape[0] - 1 ) and\
                ( ( ( proc_df_hour <= lsf_df_end_hour ) and\
                 ( proc_df_date == lsf_df_end_date ) ) or (proc_df_date < lsf_df_end_date) ):
            
            proc_df_current_line_no +=1
            proc_df_hour = proc_at_time_df.loc[proc_df_current_line_no,'time_py'].hour
            proc_df_date = proc_at_time_df.loc[proc_df_current_line_no,'date_py']
            proc_at_time_df.loc[proc_df_current_line_no,'processes_running'] += 1
    if (i % 1000 == 0):
        print(i)
```

#### Add how many processes ended and started at a time

```python
proc_at_time_df['proc_ended_this_hour'] = np.zeros(proc_at_time_df.shape[0])
proc_at_time_df['proc_started_this_hour'] = np.zeros(proc_at_time_df.shape[0])
i = 0
for lsf_current_id_process in (lsf_df['id process'].unique()):
    if (i % 1000 ==0):
        print(i)
    i+=1
    
    lsf_current_line = lsf_df[lsf_df['id process']==lsf_current_id_process].iloc[0]
#     print(lsf_current_line)
#     print('\n')
    
    #find first line of date and end_time+1 in proc_at_time_df 
    #which are the same as in lsf
    #add +1 process that ended during the reported hour
    lsf_df_start_hour = lsf_current_line.start_time_py.hour
    lsf_df_start_date = lsf_current_line.start_date_py
    
    if (lsf_df_start_date <= proc_at_time_df.iloc[len(proc_at_time_df)-1].date_py):
        if ( lsf_df_start_hour != 23 ):
            proc_df_current_line_no = proc_at_time_df[ (proc_at_time_df.date_py == lsf_df_start_date) \
                                                  & (proc_at_time_df.time_py > datetime.time(lsf_df_start_hour,0,0)) ].index[0]
        else:
            proc_df_current_line_no = proc_at_time_df[ (proc_at_time_df.date_py > lsf_df_start_date ) \
                                                  & (proc_at_time_df.time_py == datetime.time(0,0,0)) ].index[0]
        if ( proc_df_current_line_no < proc_at_time_df.shape[0] ):
#             print( proc_at_time_df.iloc[proc_df_current_line_no] )
            proc_at_time_df.loc[proc_df_current_line_no,'proc_started_this_hour'] += 1
    
    
    lsf_df_end_hour = lsf_current_line.end_time_py.hour
    lsf_df_end_date = lsf_current_line.end_date_py
    
    if (lsf_df_end_date <= proc_at_time_df.iloc[len(proc_at_time_df)-1].date_py):
        if ( lsf_df_end_hour != 23 ):
            proc_df_current_line_no = proc_at_time_df[ (proc_at_time_df.date_py == lsf_df_end_date) \
                                                  & (proc_at_time_df.time_py > datetime.time(lsf_df_end_hour,0,0)) ].index[0]
        else:
            proc_df_current_line_no = proc_at_time_df[ (proc_at_time_df.date_py > lsf_df_end_date) \
                                                  & (proc_at_time_df.time_py == datetime.time(0,0,0)) ].index[0]
        if ( proc_df_current_line_no < proc_at_time_df.shape[0] ):
#             print( proc_at_time_df.iloc[proc_df_current_line_no] )
            proc_at_time_df.loc[proc_df_current_line_no,'proc_ended_this_hour'] += 1
    else:
        continue
```

#### Add number of cores: avg, min, max at a time

```python
# Add column avg, min, max no of cores during this hour + add_counter
# First id_process in unique -> add min for this process in min, max in max, first mean to mean col
# Then go through id_proc and add the same from starting hour until ending hour including
# Watch out for exceptions: end date out of range

proc_at_time_df['cores_avg'] = np.zeros(proc_at_time_df.shape[0])
proc_at_time_df['cores_min'] = lsf_df['# core '].max()
proc_at_time_df['cores_max'] = lsf_df['# core '].min()
# proc_at_time_df['count'] = np.zeros(proc_at_time_df.shape[0])
i = 0
for lsf_current_id_process in (lsf_df['id process'].unique()):
#     print(i)
    i+=1
    lsf_current_line = lsf_df[lsf_df['id process']==lsf_current_id_process]
    
    #stats
    min_core = lsf_current_line['# core '].min()
    max_core = lsf_current_line['# core '].max()
    mean_core = lsf_current_line['# core '].mean()
    
    lsf_current_line = lsf_df[lsf_df['id process']==lsf_current_id_process].iloc[0]
    
#     print(lsf_current_line)
#     print('\n')
    
    #find first line of date and start_time+1 in proc_at_time_df 
    #which are the same as in lsf
    #add +1 process during the reported hour
    #go through time in proc_df 
    #while date<=end_date and end_time.hour (in lsf_df) +1 <= time in proc_df
    
    #did not check zero case 
    lsf_df_start_hour = lsf_current_line.start_time_py.hour
    lsf_df_end_hour = lsf_current_line.end_time_py.hour
    lsf_df_end_date = lsf_current_line.end_date_py
    
    if ( lsf_df_start_hour != 23 ):
        proc_df_current_line_no = proc_at_time_df[ (proc_at_time_df.date_py == lsf_current_line.start_date_py) \
                                              & (proc_at_time_df.time_py > datetime.time(lsf_df_start_hour,0,0)) ].index[0]
    else:
        proc_df_current_line_no = proc_at_time_df[ (proc_at_time_df.date_py > lsf_current_line.start_date_py) \
                                              & (proc_at_time_df.time_py == datetime.time(0,0,0)) ].index[0]
#     print( proc_at_time_df.iloc[proc_df_current_line_no] )
    
    proc_df_hour = proc_at_time_df.loc[proc_df_current_line_no,'time_py'].hour
#     print(proc_df_hour)

    proc_df_date = proc_at_time_df.loc[proc_df_current_line_no,'date_py']
#     print(proc_df_date)
    
    proc_at_time_df.loc[proc_df_current_line_no,'cores_avg'] += mean_core
#     proc_at_time_df.loc[proc_df_current_line_no,'count'] += 1
    if (proc_at_time_df.loc[proc_df_current_line_no,'cores_min'] > min_core):
        proc_at_time_df.loc[proc_df_current_line_no,'cores_min'] = min_core
    if (proc_at_time_df.loc[proc_df_current_line_no,'cores_max'] < max_core):
        proc_at_time_df.loc[proc_df_current_line_no,'cores_max'] = max_core
    
    
    if not( (lsf_df_start_hour == lsf_df_end_hour) and (lsf_current_line.start_date_py == lsf_df_end_date) ):
        while ( proc_df_current_line_no < proc_at_time_df.shape[0] - 1 ) and\
                ( ( ( proc_df_hour <= lsf_df_end_hour ) and\
                   ( proc_df_date == lsf_df_end_date ) ) or (proc_df_date < lsf_df_end_date) ):
            proc_df_current_line_no += 1
            proc_df_hour = proc_at_time_df.loc[proc_df_current_line_no,'time_py'].hour
            proc_df_date = proc_at_time_df.loc[proc_df_current_line_no,'date_py']

            proc_at_time_df.loc[proc_df_current_line_no,'cores_avg'] += mean_core
#             proc_at_time_df.loc[proc_df_current_line_no,'count'] += 1
            if (proc_at_time_df.loc[proc_df_current_line_no,'cores_min'] > min_core):
                proc_at_time_df.loc[proc_df_current_line_no,'cores_min'] = min_core
            if (proc_at_time_df.loc[proc_df_current_line_no,'cores_max'] < max_core):
                proc_at_time_df.loc[proc_df_current_line_no,'cores_max'] = max_core
    if (i % 1000 == 0):
        print(i)
proc_at_time_df['cores_avg'] = proc_at_time_df['cores_avg']/proc_at_time_df['processes_running']
```
